let's find recomendations.
The user should insert 3 movies and the algorithm will get the movies closest recomended to the selected ones. I took tre randomly selected movies, with the ID 18,19,20 as my initial imput. If more movies are located in the same cluster, that cluster has an increased importance. 

```python
#let's put in 3 of my favourite movies of all time: 8 1/2 from fellini (ID 1251), ducksoup from the Marx Brothers (ID 1256), Kill Bill from Tarantino (ID 6874)
pd_find_next_movie=pd_final.loc[1251:1251].append(pd_final.loc[1256:1256]).append(pd_final.loc[6874:6874])
```

```python
pd_find_next_movie
```

```python
#let's see to whitch cluster they are related
pd_find_next_movie[['K_cluster','Density_cluster']]
```

```python
#let's count the freequency of the cluster in the selected movies. In other words, let's see if one or more movies have the same cluster
list_of_density_clusters=[]
for cluster in pd_find_next_movie['Density_cluster']:
    list_of_clusster_repetition_value=[]
    reiteration_counter_1=0
    #let's check each cluster with the other clusters. let's take the first cluster and compar it to the next one. Append the result and start over with the next cluster
    while reiteration_counter_1< len(pd_find_next_movie['Density_cluster']):
        reiteration_counter_2=0
        #let's see if the cluster of one movie is the same of the others
        if np.array(cluster).round(6)==pd_find_next_movie['Density_cluster'].iloc[reiteration_counter_1].round(6):
            reiteration_counter_2+=1
        #let's append the result to the list
        list_of_clusster_repetition_value.append(reiteration_counter_2)
        reiteration_counter_1+=1
    list_of_density_clusters.append(sum(list_of_clusster_repetition_value))
```

```python
list_of_density_clusters
```

```python
#let's count the freequency of the cluster in the selected movies. In other words, let's see if one or more movies have the same cluster
list_of_K_clusters=[]
for cluster in pd_find_next_movie['K_cluster']:
    list_of_clusster_repetition_value=[]
    reiteration_counter_1=0
    #let's check each cluster with the other clusters. let's take the first cluster and compar it to the next one. Append the result and start over with the next cluster
    while reiteration_counter_1< len(pd_find_next_movie['K_cluster']):
        reiteration_counter_2=0
        #let's see if the cluster of one movie is the same of the others
        if np.array(cluster).round(6)==pd_find_next_movie['K_cluster'].iloc[reiteration_counter_1].round(6):
            reiteration_counter_2+=1
        #let's append the result to the list
        list_of_clusster_repetition_value.append(reiteration_counter_2)
        reiteration_counter_1+=1
    list_of_K_clusters.append(sum(list_of_clusster_repetition_value))
```

```python
list_of_K_clusters
```

```python
'''we previously calculated to each point in the dataset the 10 closest ones. 
Following we calculated the occurance of each cluster and are going to use that occurance to recalculate the distance previously calculated.
The distance is going to be divided by the sum of occurance of each cluster. for example, if a movie 1 is part of the same K mean cluster of movie 2 and part of the density cluster of movie 3, he is going to have value 2 for both clusters. Therefor the distances between the nearest points, ie the next 10 movie to recomend, is going to be divided by 2+2=4 instead of 1+1=2 (which is the standartd division).'''
weighted_movie_1_distance=(pd_find_next_movie.iloc[0:1,-10:]/(list_of_K_clusters[0]+list_of_density_clusters[0]))
weighted_movie_2_distance=(pd_find_next_movie.iloc[1:2,-10:]/(list_of_K_clusters[1]+list_of_density_clusters[1]))
weighted_movie_3_distance=(pd_find_next_movie.iloc[2:3,-10:]/(list_of_K_clusters[2]+list_of_density_clusters[2]))
```

```python
pd_find_next_movie.iloc[0:1,-10:]
```

```python
#let's associate the new distances to the movies. Those are the movies that the user set up initially. The 3 movies the algoritmus uses to recomend new ones
new_distance=(weighted_movie_1_distance.append(weighted_movie_2_distance)).append(weighted_movie_3_distance)
```
